```python
from sane import SANEAlgorithm
from sane import Cancer1Dataset
import random
import matplotlib.pyplot as plt
```

```python
random.seed(1)
```

```python
algorithm_params = [
        (10, 30, 5, 5),
        (10, 30, 10, 20),
        (20, 60, 15, 50)
]
```

```python
algorithms = []
for params in algorithm_params:
        algorithms.append(SANEAlgorithm(
                blueprints_population_size=params[0],
                neuron_population_size=params[1],
                hidden_layer_size=params[2],
                connections_count=params[3]))
```

```python
for i in range(len(algorithms)):
        algorithms[i].init(-1.0, 1.0)
```

```python
cancer1_dataset = Cancer1Dataset()
x_train, y_train = cancer1_dataset.get_train_data()
```

```python
results = []
for i in range(len(algorithms)):
        result = algorithms[i].train(500, x_train, y_train)
        results.append(result)
```

```python
fitness_list = []
for i in range(len(results)):
        fitness_list.append([item[0] for item in results[i]])
```

```python
legend = []
for i in range(len(fitness_list)):
       plt.plot(fitness_list[i])
       legend.append("P" + str(i))
plt.legend(legend)
plt.show()
```
